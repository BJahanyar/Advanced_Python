```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
image = cv2.imread('low.tif')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.imshow( image , cmap='gray')
cv2.waitKey()
```

```python
hist  = cv2.calcHist ([image] , [0] , None , [255] , [0, 255])
plt.plot(hist)
```

```python
Optimized = np.zeros((image.shape[0],image.shape[1]),dtype = 'uint8')

min = np.min(image)
max = np.max(image)

for i in range(image.shape[0]):
    for j in range(image.shape[1]):
        Optimized[i,j] = 255*(image[i,j]-min)/(max-min)

plt.imshow( Optimized , cmap='gray')
cv2.imwrite('high.jpg' , Optimized)
cv2.waitKey()
```

```python
hist  = cv2.calcHist ([Optimized] , [0] , None , [255] , [0, 255])
plt.plot(hist)
```

```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import random
from numpy.linalg import inv
```

```python
N = 100
X = []
Y = []

couter = 1
while ( couter <= N ) :   
    h = random.randint(150 , 190)
    w = random.randint(60 , 100) 
    BMI = w /((h/100)**2)
    if BMI > 25 and BMI < 35:
        X.append(h)
        Y.append(w)
        couter += 1

X = np.array(X)
Y = np.array(Y)

X = X.reshape(N,1)
Y = Y.reshape(N,1)
```

```python
#model = AdalineRegressor()
#model.fit(X,Y)
#print(model.w)

w = np.matmul(np.linalg.inv(np.matmul(X.T, X)),  np.matmul(X.T, Y))
print(w)
Y_pred = np.matmul(X, w)

plt.scatter(X,Y)
plt.plot(X,Y_pred, c = 'red', lw = 2)
plt.xlabel("height")
plt.ylabel("weight")
```

```python
from sklearn import linear_model

regr = linear_model.LinearRegression()
regr.fit(X, Y)
y_pred = regr.predict(X)
plt.scatter(X,Y)
plt.plot(X, y_pred, color='yellow' , lw =2)
plt.xlabel("height")
plt.ylabel("weight")
```

```python
from scipy.stats import linregress

result  = linregress(X[:, 0], Y[:, 0])
print(result.slope)
Y_pred = result.intercept + np.matmul(X, np.array([[result.slope]]))
plt.scatter(X, Y)
plt.plot(X, y_pred, c='yellow', lw = 2)
plt.xlabel('Height')
plt.ylabel('Weight')
```

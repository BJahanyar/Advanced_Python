```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('COVID-19-time-series-clean-complete.csv')
data.head()
```

```python
data['Date'] = pd.to_datetime(data['Date'])
data.head(2)
```

```python
data['MDate'] = data['Date'].map(lambda date : 100*date.year + date.month)
data.head(2)
```

```python
data['Monthly_Case'] = data['New cases'] + data['New deaths']
data
```

```python
Monthly_revenue = data.groupby(['MY_Date'])['Monthly_Case'].sum().reset_index()
Monthly_revenue
```

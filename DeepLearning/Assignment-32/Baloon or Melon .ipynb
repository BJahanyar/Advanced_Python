```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
```

```python
N = 500
std = 1.5
#ایجا  لیستی 500 تایی از اعداد به عنوان طول عرض وزن
Balloons_width = np.random.normal(17, std , N)
Balloons_length = np.random.normal(17, std , N)
Balloons_weight = np.random.normal(0.3, 0.1, N)

Melon_width = np.random.normal(5, std, N)
Melon_length = np.random.normal(15, std, N)
Melon_weight= np.random.normal(5, std, N)

#قرار داد طول هادر یک لیست در کنار هم و الحاق عرض ها و وزن ها 
width = np.concatenate((Balloons_width, Melon_width))
length = np.concatenate((Balloons_length, Melon_length))
weight = np.concatenate((Balloons_weight, Melon_weight))

# ایجاد یک آرایه دو بعدی که سطر های 1000 است و ستون های عرض و طول و وزن 3 است. ما به عنوان 
#ویژگی که میخواهیم روی آن ها کار کنیم
# ایجاد 500 عدد صفر و 500عدد یک برای انتصاب به لیبل بالن هاو خربزه ها
X = np.array([width, length, weight]).T #تعویض جای سطر و ستون ها
#Balloons_lable = np.zeros(N, dtype = 'int')
#Melon_lable = np.ones(N, dtype = 'int')

#به هم پیوستن لیبل بالن ها و خربزه ها برای ایجاد خروجی یا تارگت
#Y = np.concatenate((Balloons_lable,Melon_lable))
```

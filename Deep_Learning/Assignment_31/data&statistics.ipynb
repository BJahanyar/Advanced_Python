```python
import numpy as np
import matplotlib.pyplot as plt
import pandas



speed = [180,120,160,80,65,75,180,100,75,75,60,45,90,180]
age = [1,3,2,5,9,6,1,4,4,5,8,10,5,1]
```

```python


#############_____M E A N_____#############

my_mean = np.mean(speed)
print(my_mean)

#############_____M E D I A N_____#############

my_median = np.median (speed)
print(my_median)

#############_____M O D E_____#############
from scipy import stats

my_mode = stats.mode(speed)
print(my_mode)

#############_____D E V I A T I O N_____#############

my_deviation = np.std (speed)
print(my_deviation)

#############_____V A R I A N C E_____#############

my_variance = np.var(speed)
print(my_variance)

#############_____P E R C E N T I L E S_____#############

my_percentile = np.percentile(speed, 50)
print(my_percentile)

```

```python

#############_____R E G R E S S I O N_____#############

slope, intercept, r, p, std_err = stats.linregress(speed,age)

def myfunction(speed):
    return slope * speed + intercept

mymodel = list (map(myfunction,speed))

print(r)
print(p)
print(std_err)
print(slope)
print(intercept)

plt.scatter(speed,age)
plt.plot(speed, mymodel)
plt.show()
```

```python
#############_____M U L T I P L E__R E G R E S S I O N_____#############
from sklearn import linear_model
from sklearn.preprocessing import StandardScaler

df = pandas.read_csv("cars.csv")

X = df[['Weight' ,'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

CO2Prediction = regr.predict([[1300,700]])

print( 'predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3 ------->', CO2Prediction )
print('\n')

#############_____C O E F F I C I E N T_____#############

coeddicient = regr.coef_

print('The coefficient value of weight and  volume against CO2 is ------->' , coeddicient)
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

df = pandas.read_csv("shows.csv")

d = {'UK': 0, 'USA': 1, 'N': 2}
df['Nationality'] = df['Nationality'].map(d)
d = {'YES': 1, 'NO': 0}
df['Go'] = df['Go'].map(d)

features = ['Age', 'Experience', 'Rank', 'Nationality']

X = df[features]
y = df['Go']

dtree = DecisionTreeClassifier()
dtree = dtree.fit(X ,y)
data = tree.export_graphviz(dtree, out_file=None , feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('mydecisiontree.png')

img = pltimg.imread('mydecisiontree.png')
imgplot = plt.imshow(img)
plt.show()
```
